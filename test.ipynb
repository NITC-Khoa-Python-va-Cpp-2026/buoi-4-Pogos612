```python
with open("input.txt", "r", encoding="utf-8") as content:
    data = content.read()
with open("copy.txt", "w", encoding="utf-8") as content2:
    content2.write(data)
```

```python
import random

with open("copy.txt", "r") as content:
    all_data = content.readlines()

fiveRandomLines = random.sample(all_data, 5)

with open("output.txt", "w", encoding= "utf-8") as content2:
    content2.writelines(fiveRandomLines)
    content2.write("Hoàng Hà Giang")
```
